# 3. Data in columns "last_review" and "reviews_per_month" is not very important for us right now, so I will leave empty cells there.  I will fill the price column with average value for the area and type of apartment.

```python
#Here I fill data based on columns 'neighbourhood'and 'room_type'
data['price'].fillna(data.groupby(['neighbourhood','room_type'])['price'].transform('mean'),inplace=True)
#Next step I fill rest of data based on columns 'neighbourhood_group'and 'room_type' for appartments with unique
#'neighbourhood' value
data['price'].fillna(data.groupby(['neighbourhood_group','room_type'])['price'].transform('mean'),inplace=True)
```

```python
#in this part we built scatter plot to see our possible outliers
sc_fig= px.scatter(data,x=data['price'],y=data['minimum_nights'])
sc_fig.show()
```

```python
#Lets count amount of outliers and clean them
price_check=data[data['price']>1500].count()
min_nights_check=data[data['minimum_nights']>100].count()
print('price_check =', price_check['id'])
print('min_nights_check =', min_nights_check['id'])
# we can remove outliers cause there are small amount of them
clean_data=data[(data['price']<=1500) & (data['minimum_nights']<=100)]
```

# 4. In this part I will analyze the relationship between price and neighbourhood group and divide it by apartment type

```python
avg_price = clean_data.groupby(['neighbourhood_group', 'room_type'])['price'].mean().reset_index()
fig = px.bar(avg_price, x='neighbourhood_group', y='price', color='room_type', barmode='group')
fig.update_layout(coloraxis_colorscale='matter')
fig.show()
```

```python
def plot_neighborhood_group(data, group_name):
  data_filtered = data[data['neighbourhood_group'] == group_name]
  avg_price_upd = data_filtered.groupby(['neighbourhood', 'room_type'])['price'].mean().reset_index()
  fig = px.bar(avg_price_upd, x='neighbourhood', y='price', color='room_type', barmode='group',title=group_name)
  fig.update_layout(coloraxis_colorscale='matter')
  fig.show()

for group_name in data['neighbourhood_group'].unique():
  plot_neighborhood_group(clean_data, group_name)
```

* On these graphs we can see price change depanding on the type of apartment and location.

```python
avg_price_map= clean_data.groupby(['neighbourhood', 'room_type','longitude','latitude'])['price'].mean().reset_index()
fig= px.scatter(avg_price_map,x=avg_price_map['longitude'],y=avg_price_map['latitude'],color=avg_price_map['price'])
fig.show()
```

* Here we can see our heatmap for the average price of apartment

# 5. Now we will find out the top 5 landlords and the amount of apartments they have

```python
top_5_landlords=data['host_id'].value_counts()[:5]
print(top_5_landlords)
top_5_landlords_ids = top_5_landlords.index.to_list()
```

```python
top_landlord_data= data[data['host_id'].isin(top_5_landlords_ids)]
avg_price_top_5 = (top_landlord_data.groupby(['room_type', 'neighbourhood_group', 'host_name'])[['price','host_id']].mean().reset_index())
avg_price_top_5.rename(columns={'price': 'avg_price'}, inplace=True)
count_by_user_type_neighbourhood = top_landlord_data.groupby(['room_type', 'neighbourhood_group', 'host_id']).size().reset_index(name='count')
combined_data = avg_price_top_5.merge(count_by_user_type_neighbourhood, on=['room_type', 'neighbourhood_group', 'host_id'], how='left')
fig = px.bar(
    combined_data, x='neighbourhood_group', color='host_name',
    y='avg_price',
    barmode='group',
    facet_col='room_type',
    hover_name='host_id',
    hover_data=['neighbourhood_group', 'room_type', 'count','avg_price']
)
fig.update_layout(
    title='Apartment Distribution (Top Landlords): Count by Host per Neighbourhood Group and Room Type',
    coloraxis_colorscale='matter'
)
fig.show()
```

* as we can see on the graphs, the most common apartment type are private, and the price is more or less the same regardless of the landlord.

```python
top_landlord_data= data[data['host_id'].isin(top_5_landlords_ids)]
avg_price_top_5 = (top_landlord_data.groupby(['room_type', 'neighbourhood_group', 'host_name'])[['price','host_id']].mean().reset_index())
avg_price_top_5.rename(columns={'price': 'avg_price'}, inplace=True)
count_by_user_type_neighbourhood = top_landlord_data.groupby(['room_type', 'neighbourhood_group', 'host_id']).size().reset_index(name='count')
combined_data = avg_price_top_5.merge(count_by_user_type_neighbourhood, on=['room_type', 'neighbourhood_group', 'host_id'], how='left')
fig = px.scatter(
    combined_data, x='neighbourhood_group', color='host_name',
    y='count',
    size='avg_price',
    facet_col='room_type',
    hover_name='host_id',
    hover_data=['neighbourhood_group', 'room_type', 'count','avg_price']
)
fig.update_layout(
    title='Apartment Distribution (Top Landlords): Count by Host per Neighbourhood Group and Room Type',
    coloraxis_colorscale='matter'
)
fig.show()
```

* On these graphs we can see the price distribution and the amount of apartments between top 5 biggest landlords.

# 6. We can also analyze the apartments by the amount of reviews and see their description and location.


```python
reviews_data=data[data['number_of_reviews']>0]
reviews_data['number_of_reviews'].value_counts()
```

```python
top_reviewed_apartments = reviews_data.nlargest(100, 'number_of_reviews')
# fig = px.scatter_mapbox(
#     top_reviewed_apartments,
#     lat='latitude',
#     lon='longitude',
#     hover_name='name',
#     hover_data=['price', 'room_type', 'host_name'],
#     color='number_of_reviews',
#     zoom=10,
#     title='Top 100 Reviewed Apartments (Price, Type, Host on Hover)'
# )
# fig.show()

url = 'https://raw.githubusercontent.com/plotly/datasets/master/2014_us_cities.csv'
city_data = pd.read_csv(url)
scale = 10
fig = Figure()

neighbourhoods = top_reviewed_apartments['neighbourhood'].tolist()
data_neighbourhoods=[clean_data['neighbourhood'].unique()]
appartments=top_reviewed_apartments['id'].unique()
for n in appartments:
  df_sub = top_reviewed_apartments[top_reviewed_apartments['id'] == n]
  name=df_sub['name'].iloc[0]
  price=df_sub['price'].iloc[0]
  room_type=df_sub['room_type'].iloc[0]
  trace_name=f"{name}/nPrice:{price}/n{room_type}"
  fig.add_trace(go.Scattergeo(
      locationmode='USA-states',
      lon=df_sub['longitude'],
      lat=df_sub['latitude'],
      text=trace_name,
      mode='markers',
      marker=dict(
          size=top_reviewed_apartments['price']/scale,
          line_color='rgb(40,40,40)',
          line_width=0.5,
          sizemode='area'
      )
  ))
fig.update_geos(fitbounds="locations")
# fig.update_traces(name=name)
fig.update_layout(
    title_text='Top 100 Reviewed Apartments (Price, Type, Host on Hover)',
    showlegend=True,
    geo=dict(
        scope='usa',
        landcolor='rgb(217, 217, 217)',
    )
)

fig.show()
```
